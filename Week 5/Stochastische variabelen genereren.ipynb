Voor mijn random nummer generator heb ik voor een Linear Congruential Generator(LCG) gekozen.

De LCG gebruik de formule $X_{n+1} = aX_n + c\mod m $ waarbij $a,c,m âˆˆ \mathbb{Z}_+$ vaste parameters zijn. Hierdoor worden de gegenereerde getallen bepaald door de seed $X_0$.

```python
def lcg(n, x0, a, c, m):
    prs = [];
    xn = x0;
    for i in range (n):
        xn = (a*xn + c) % m;
        prs.append (xn);
    return prs;
print(lcg(10, 0, a=5, c=1, m=8));
```

De bovenstaande getallen lijken willekeurig, maar dit blijkt toch anders te zijn als ze in een plot van 100 getallen worden gezet. 

```python
import matplotlib;
import matplotlib.pyplot as plt; # for plotting

(x0, a, c, m) = (0, 5, 1, 8);  
n = 100;
plt.plot (lcg (n, x0, a, c, m));
plt.show ();
```

Je ziet telkens een terugkomend patroon van getallen dat zich om de $x$ aantal getallen herhaald. Dit komt omdat de parameters $a,c$ en $m$ van groots belang zijn voor de (psuedo)willekeurigheid. Als we nu de waarden $a = 1103515245, c = 12345, m = 2**32$ dan krijgen we dit als uitkomst.

```python
(a, c, m) = (1103515245, 12345, 2**32);
plt.plot (lcg (n, x0, a, c, m))
plt.show ();
```

```python
def sort(sequence):
    sequence.sort()
    return sequence
n = 10**4
plt.hist(sort(lcg (n, x0, a, c, m)))
```

Met bovenstaande plots is te zien dat de verdeling gelijk is. Zoals in de onderste plot te zien is dat deze willekeurige getallen toch normaal verdeelt zijn van elkaar.
